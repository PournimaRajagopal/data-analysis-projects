### Setting Up Your Data

Before you dive into data visualizations, you need to set up your dataframe and make sure that it is cleaned and ready to go.

```python
# import all librarys and modules needed
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style
# setting style for graphs
style.use('ggplot')
plt.rcParams['figure.figsize'] = (20,10)
```

```python
# import and clean data
df = pd.read_csv("./Popular_Spotify_Songs.csv", encoding='latin-1')

df.isna().sum()
df.describe()
```

### Creating Visualizations

Now that the data is ready, you can use the dataset to create a number of different visualizations. You do not need to make the visualizations perfect; this is just for practice.

#### Line Chart

```python
# Single line chart
df['danceability_%'].plot(kind='line')
```

```python
# Multiple Line chart 
df.plot(kind='line',subplots = True)
```

#### Bar Chart

```python
# Vertical bar chart
df.plot(kind='bar')
```

```python
# Horizontal bar chart
plt.barh(df['artist_count'],df['artist(s)_name'])
plt.show()
```

```python
# Grouped bar chart
df.head(20).plot(kind='bar',x='artist_count',stacked = True)
```

#### Scatter Plot

```python
# scatter plot
df.plot(kind='scatter',x='in_apple_charts',y='in_spotify_charts')
```

#### Histogram

```python
# histogram
sb.histplot(df, bins = 6)
```

#### Pie Chart

```python
# pie chart
df.plot.pie(y='released_year')
```
